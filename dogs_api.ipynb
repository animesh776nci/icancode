```python
import pymongo
import json
from pymongo import MongoClient
client = MongoClient()
db = client['proj_1']
```

```python
dog_db1 = db['animals']
import requests
def petFinderAPI(page):
    data = {
      'grant_type': 'client_credentials',
      'client_id': 'LBgEzxajv0GfJRPg7lpCiXcVNqBJjUVg7WKqVyQjIVzB5eGqXy',
      'client_secret': '3QBkUvDPRmmphHPTvI1ZH2kDG4yhS4EY5Muu0Y7r'
    }

    token = requests.post('https://api.petfinder.com/v2/oauth2/token', data=data)

    headers = {
        'Authorization': 'Bearer '+(json.loads(token.content)['access_token']),
    }
    params = (
        ('type', 'dog'),
        ('page', page),
    )

    response = requests.get('https://api.petfinder.com/v2/animals', headers=headers, params=params)
    return json.loads(response.content)

for j in range(1,101):
    for i in petFinderAPI(j)["animals"]:
        dog_db1.insert_one(i)
client.close()
```
